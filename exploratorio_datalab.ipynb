```python
from collections import Counter
import re
import pandas as pd

# Concatenar 'review_title' y 'review_content'
amazon_unificado['title_content'] = amazon_unificado['review_title'] + " " + amazon_unificado['review_content']

# Función para limpiar y contar palabras
def contar_palabras(texto):
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar caracteres especiales y dividir en palabras
    palabras = re.findall(r'\b\w{4,}\b', texto)
    # Contar palabras
    return Counter(palabras)

# Aplicar la función a toda la columna 'title_content'
conteo_palabras = amazon_unificado['title_content'].apply(contar_palabras)

# Sumar todos los contadores para obtener las 25 palabras más comunes
suma_conteos = sum(conteo_palabras, Counter())
palabras_comunes = [palabra for palabra, _ in suma_conteos.most_common(25)]

# Crear columnas para cada una de las 25 palabras más comunes
for palabra in palabras_comunes:
    amazon_unificado[palabra] = amazon_unificado['title_content'].apply(lambda x: x.lower().split().count(palabra))

# Seleccionar solo las columnas de interés
amazon_resultado = amazon_unificado[['title_content'] + palabras_comunes]

# Exportar el DataFrame resultante a un archivo HTML
amazon_resultado.to_html('amazon_resultado.html', index=False)

# Convertir el conteo total a DataFrame para un mejor display
df_suma_conteos = pd.DataFrame(suma_conteos.items(), columns=['Palabra', 'Conteo']).sort_values(by='Conteo', ascending=False)

# Exportar el DataFrame de conteo total a un archivo HTML
df_suma_conteos.to_html('conteo_total.html', index=False)

print("Los archivos se han exportado como 'amazon_resultado.html' y 'conteo_total.html'.")
```

```python
import pandas as pd
import numpy as np
from IPython.display import display

# DataFrame para almacenar datos numéricos
amazon_numericos = pd.DataFrame()

# Recorrer cada columna y validar tipo de dato
for column in amazon_unificado.columns:
    try:
        # Verificar si la columna tiene valores numéricos
        if amazon_unificado[column].apply(lambda x: str(x).replace('.', '', 1).isdigit()).all():
            # Convertir a tipo numérico
            amazon_unificado[column] = pd.to_numeric(amazon_unificado[column], errors='coerce')
            amazon_numericos[column] = amazon_unificado[column]
        else:
            # Si es string, mantener en el DataFrame original
            amazon_unificado[column] = amazon_unificado[column].astype(str)
    except Exception as e:
        print(f"Error procesando la columna '{column}': {e}")

# Mensaje si el DataFrame de datos numéricos está vacío
if amazon_numericos.empty:
    print("No se encontraron columnas numéricas en el DataFrame.")

# Medidas de tendencia central y dispersión para datos numéricos
try:
    if not amazon_numericos.empty:
        # Descripción básica
        tendencia_central = amazon_numericos.describe().T
        
        # Agregar medidas adicionales
        tendencia_central['varianza'] = amazon_numericos.var()
        tendencia_central['desviación estándar'] = amazon_numericos.std()
        tendencia_central['moda'] = amazon_numericos.mode().iloc[0]  # Moda puede ser multi-valor
        tendencia_central['mediana'] = amazon_numericos.median()
        tendencia_central['rango intercuartil'] = tendencia_central['75%'] - tendencia_central['25%']
        
        # Renombrar columnas para un formato más claro
        tendencia_central.rename(columns={
            'mean': 'mean/promedio',
            'std': 'std/desviación',
            'min': 'min/minimo',
            '25%': '25%/primer cuartil',
            '50%': '50%/mediana',
            '75%': '75%/tercer cuartil',
            'max': 'max/maximo'
        }, inplace=True)
        
        # Seleccionar solo las columnas relevantes
        tendencia_central = tendencia_central[['mean/promedio', 'std/desviación', 'min/minimo', '25%/primer cuartil', '50%/mediana', '75%/tercer cuartil', 'max/maximo', 'varianza', 'desviación estándar', 'moda', 'rango intercuartil']]
        
        # Resetear índices para que los nombres de las columnas sean más legibles
        tendencia_central.reset_index(inplace=True)
        tendencia_central.rename(columns={'index': 'Columna'}, inplace=True)
        
        # Mostrar tabla con display
        display(tendencia_central)
    else:
        print("No se pueden calcular medidas de tendencia central porque no hay columnas numéricas.")
except Exception as e:
    print(f"Error calculando las medidas de tendencia central: {e}")

# Verificación de errores en columnas string
for column in amazon_unificado.columns:
    if amazon_unificado[column].dtype == 'object':
        if any(amazon_unificado[column].str.contains('[^0-9.]', regex=True)):
            print(f"Advertencia: La columna '{column}' contiene caracteres no numéricos.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot

# Asegurarse de que las columnas 'rating' y 'discount_percentage' sean numéricas
amazon_unificado['rating'] = pd.to_numeric(amazon_unificado['rating'], errors='coerce')
amazon_unificado['discount_percentage'] = pd.to_numeric(amazon_unificado['discount_percentage'], errors='coerce')

# Paso 1: Identificación del Tipo de Variables
rating_is_numeric = pd.api.types.is_numeric_dtype(amazon_unificado['rating'])
discount_is_numeric = pd.api.types.is_numeric_dtype(amazon_unificado['discount_percentage'])

if rating_is_numeric and discount_is_numeric:
    variable_type = "Ambas variables son continuas."
    print(f"Variables: 'rating' y 'discount_percentage' son continuas y numéricas.")

    # Paso 2: Evaluación de la Distribución de los Datos
    print("\nPaso 2: Evaluación de la Distribución de los Datos")

    # Test de normalidad para 'rating'
    rating_shapiro = stats.shapiro(amazon_unificado['rating'].dropna())
    print(f"Shapiro-Wilk Test para 'rating': p-valor = {rating_shapiro.pvalue:.4f}")

    # Test de normalidad para 'discount_percentage'
    discount_shapiro = stats.shapiro(amazon_unificado['discount_percentage'].dropna())
    print(f"Shapiro-Wilk Test para 'discount_percentage': p-valor = {discount_shapiro.pvalue:.4f}")

    # Visualización de la distribución
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(amazon_unificado['rating'].dropna(), kde=True, color='blue')
    plt.title('Distribución de Rating')

    plt.subplot(1, 2, 2)
    sns.histplot(amazon_unificado['discount_percentage'].dropna(), kde=True, color='green')
    plt.title('Distribución de Discount Percentage')
    plt.show()

    # Q-Q Plots para ver la normalidad
    qqplot(amazon_unificado['rating'].dropna(), line='s')
    plt.title('Q-Q Plot para Rating')
    plt.show()

    qqplot(amazon_unificado['discount_percentage'].dropna(), line='s')
    plt.title('Q-Q Plot para Discount Percentage')
    plt.show()

    # Paso 3: Selección del Test Estadístico
    print("\nPaso 3: Selección del Test Estadístico")
    if rating_shapiro.pvalue > 0.05 and discount_shapiro.pvalue > 0.05:
        # Ambas variables siguen una distribución normal
        print("Ambas variables siguen una distribución normal. Se utilizará el Test de Correlación de Pearson y el Test de Correlación de Kendall.")
        corr_test1 = "pearson"
        correlation1, p_value1 = stats.pearsonr(amazon_unificado['rating'].dropna(), amazon_unificado['discount_percentage'].dropna())

        # Test adicional: Kendall para comparación
        corr_test2 = "kendall"
        correlation2, p_value2 = stats.kendalltau(amazon_unificado['rating'].dropna(), amazon_unificado['discount_percentage'].dropna())
        
    else:
        # Al menos una variable no sigue una distribución normal
        print("Al menos una variable no sigue una distribución normal. Se utilizará el Test de Correlación de Spearman y el Test de Correlación de Kendall.")
        corr_test1 = "spearman"
        correlation1, p_value1 = stats.spearmanr(amazon_unificado['rating'].dropna(), amazon_unificado['discount_percentage'].dropna())

        # Test adicional: Kendall para comparación
        corr_test2 = "kendall"
        correlation2, p_value2 = stats.kendalltau(amazon_unificado['rating'].dropna(), amazon_unificado['discount_percentage'].dropna())

    # Paso 4: Verificación de Supuestos y Resultados
    print("\nPaso 4: Verificación de Supuestos y Resultados")

    def describe_correlation(correlation):
        if correlation == 1:
            return "perfecta y positiva"
        elif correlation == -1:
            return "perfecta y negativa"
        elif 0.7 < correlation < 1:
            return "fuerte y positiva"
        elif -1 < correlation < -0.7:
            return "fuerte y negativa"
        elif 0.4 < correlation <= 0.7:
            return "moderada y positiva"
        elif -0.7 <= correlation < -0.4:
            return "moderada y negativa"
        elif 0.2 < correlation <= 0.4:
            return "débil y positiva"
        elif -0.4 <= correlation < -0.2:
            return "débil y negativa"
        else:
            return "muy débil"

    # Resultados del Test 1
    print(f"Test 1 Seleccionado: {corr_test1.capitalize()} Correlation Test")
    print(f"Coeficiente de Correlación: {correlation1:.4f} ({describe_correlation(correlation1)})")
    print(f"p-valor: {p_value1:.4f}")
    if p_value1 < 0.05:
        print(f"Existe una correlación estadísticamente significativa entre 'rating' y 'discount_percentage' según el test de {corr_test1.capitalize()}.")
    else:
        print(f"No se encontró una correlación estadísticamente significativa entre 'rating' y 'discount_percentage' según el test de {corr_test1.capitalize()}.")

    # Resultados del Test 2
    print(f"\nTest 2 Seleccionado: {corr_test2.capitalize()} Correlation Test")
    print(f"Coeficiente de Correlación: {correlation2:.4f} ({describe_correlation(correlation2)})")
    print(f"p-valor: {p_value2:.4f}")
    if p_value2 < 0.05:
        print(f"Existe una correlación estadísticamente significativa entre 'rating' y 'discount_percentage' según el test de {corr_test2.capitalize()}.")
    else:
        print(f"No se encontró una correlación estadísticamente significativa entre 'rating' y 'discount_percentage' según el test de {corr_test2.capitalize()}.")

else:
    print("Una o ambas variables no son numéricas. No se puede realizar la correlación.")
```
